```python
import requests #módulo para coletar uma página web
import pandas as pd
import numpy as np
from bs4 import BeautifulSoup   #biblioteca para extrair dados de arquivos HTML e XML
import re #regex
```

```python
#função para buscar os dados
def buscarDadosOlx(pages=40):  #pages = números de páginas que será feito a raspagaem
    
    #pagina da olx com os filtros
    url_pag = 'https://rn.olx.com.br/?q=apartamentos%20para%20vender%20em%20natal'
    
    #dicionario para salvar os dados
    dic_apartamento = {'Título apartamento':[], 'Preço do apartamento':[], 'Quantidade de quarto':[], 'Area do apartamento':[],
                       'Taxa do condomínio':[], 'Vagas de garagem':[], 'Endereco do apartamento':[], 'Quantidade de suítes':[],
                       'Quantidade de banheiro':[]}
                   
    #for para acessar todas as paginas desejadas
    for x in range(0,pages):
        if x ==0:
            print(url_pag)
        else:
            y=x+1
            url_pag = 'https://rn.olx.com.br/?o='+str(y)+'&q=apartamentos%20para%20vender%20em%20natal'
            #print(url_pag)
        
        
        #User do google
        headers={'User-Agent':"Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/103.0.0.0 Safari/537.36"}
        
        #coleta dos dados da url
        site = requests.get(url_pag, headers=headers)
        #print(site) #out: [200] - sucesso 
        
        #Arvore de análise para a página
        soup = BeautifulSoup(site.content, 'html.parser')
        
        #Isso irá retornar todas as instâncias dentro do documento
        itens = soup.find_all('li', {'class':"sc-1fcmfeb-2 fvbmlV"})
        #print(len(itens)) #quantidade de itens analisados por pagina
       
        #for para acessar os detalhes de cada item
        for item in itens:
            
            #retorna o titulo do anuncio
            try:
                titulo = item.findAll("h2")[0].contents[0]
            
            except:
                
                titulo = " "
            
            
            #retorna o preço do apartamento
            try:
                precoAp = item.findAll("span", class_="m7nrfa-0 eJCbzj sc-fzsDOv kHeyHD")[0].contents[0]
            
            except:
                
                precoAp = " "
            
            #retorna a quantidade de quartos
            try:
                quantidadeQuarto = item.findAll("span", class_="sc-1ftm7qz-0 itsfPe sc-bdVaJa bxVNCd")[1].contents[0]
            except:
                
                quantidadeQuarto = " "
                
            #retorna a area do apartamento    
            try:
                areaAp = item.findAll("span", class_="sc-1ftm7qz-0 itsfPe sc-bdVaJa bxVNCd")[2].contents[0]
            except:
                
                areaAp = " "
                
            
            #retorna a taxa de condominio
            try:
                taxaCondominio = item.findAll("span", class_="sc-1ftm7qz-0 itsfPe sc-bdVaJa bxVNCd")[3].contents[0]
                taxaCondominio =  taxaCondominio.replace("Condomínio: " ,  "")
            except:
                
                taxaCondominio = " "
                
            
            #retorna a quantidade de vagas na garagem
            try:
                vagasGaragem = item.findAll("span", class_="sc-1ftm7qz-0 itsfPe sc-bdVaJa bxVNCd")[4].contents[0]
            except:
                
                vagasGaragem = " "
                
            
            #retorna o endereço do apartamento
            try:
                enderecoAp = item.findAll("span", class_="sc-1c3ysll-1 iDvjkv sc-fzsDOv dTHJIA")[0].contents[0]
                enderecoAp = enderecoAp.replace("- DDD 84" , "")
            except:
                
                enderecoAp = " "
            
            
            try:     
                #acessar alguns detalhes do item
                url_item = item.find("a")["href"]
                #print(url_item)
                site_item = requests.get(url_item, headers=headers)
                soup_item = BeautifulSoup(site_item.content, 'html.parser')
                
                
                #retorna a quantidade de banheiro
                quantidadeBanheiro = soup_item.findAll("dd", class_="sc-1f2ug0x-1 ljYeKO sc-ifAKCX kaNiaQ")[4].contents[0]
            except:
                
                quantidadeBanheiro = " "
            
        
            #retorna a quantidade de suite
            try:
                text= soup_item.findAll("span", class_="sc-1sj3nln-1 eOSweo sc-ifAKCX cmFKIN")[0].contents[0]
                #print(text)
                ache = re.findall( '\d' '\s' r'suíte', text)  #regex para encontrar no texto a quantidade de suite
                #print(ache)
                quantidadeSuite = ache[0]
            except:
                quantidadeSuite = " "
                
               
            
                #print("\n")
                #print('TÍTULO = ',titulo)
                #print('PREÇO DO AP = ',precoAp)
                #print('QUANTIDADE DE QUARTOS = ',quantidadeQuarto)
                #print('AREA DO AP = ',areaAp)
                #print('TAXA DO CONDOMINIO = ',taxaCondominio)
                #print('QUANTIDADE DE VAGAS NA GARAGEM = ',vagasGaragem)
                #print('ENDEREÇO DO AP = ',enderecoAp)
                #print('QUANTIDADE DE BANHEIRO = ',quantidadeBanheiro)
                #print('QUANTIDADE DE SUÍTES = ',quantidadeSuite)
                #print("\n")
                
                
               #guardando os valores
                dic_apartamento['Título apartamento'].append(titulo)
                dic_apartamento['Preço do apartamento'].append(precoAp)
                dic_apartamento['Quantidade de quarto'].append(quantidadeQuarto)
                dic_apartamento['Area do apartamento'].append(areaAp)
                dic_apartamento['Taxa do condomínio'].append(taxaCondominio)
                dic_apartamento['Vagas de garagem'].append(vagasGaragem)
                dic_apartamento['Endereco do apartamento'].append(enderecoAp)
                dic_apartamento['Quantidade de banheiro'].append(quantidadeBanheiro)
                dic_apartamento['Quantidade de suítes'].append(quantidadeSuite)
                
            
                
            
  
        #print(dic_apartamento)
        #transformando o dicionario em um dataframe para transformar no arq csv
        df = pd.DataFrame(dic_apartamento)
        df.to_csv('mercado_imobiliario_ws.csv', encoding='utf-8', sep='|')
```

```python
buscarDadosOlx()
```
